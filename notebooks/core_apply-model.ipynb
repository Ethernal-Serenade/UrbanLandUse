### Apply model
We can apply a model in at least two ways. One, we can do a limited application of the model, and classify only those pixels for which we have ground-truth in order to quantitatively assess the quality of the model's predictions, anywhere we have ground-truth. Two, we can use it to classify every pixel in every tile in order to generate a comprehensive LULC raster, for any city for which we have imagery (and any other necessary remote sensing inputs). And of course, we can do both.

### Use model to generate performance statistics

Assign application-specific variables

```python
label_suffix = 'aue'

cats_map = {}
cats_map[0] = 0
cats_map[1] = 1
cats_map[2] = 4
cats_map[3] = 4
cats_map[4] = 4
cats_map[5] = 4
cats_map[6] = 6

notes = 'test use of new scorecard functionality'
```

Load the samples of interest. Just as when training a model, this dataset can be constructed from one or more images of one or more cities.

```python
place_images = {}
place_images['hyderabad'] = ['A']

X_train, Y_train_raw, X_valid, Y_valid_raw = util_workflow.load_datasets(place_images, data_root, label_suffix, stack_label, window)
```

Prepare data for model ingestion. This process is contingent on what model is being applied and what taxonomy that model employs.

```python
X_train_scaled, X_valid_scaled, scaler = util_ml.scale_learning_data(X_train, X_valid)
```

```python
Y_train = Y_train_raw.copy()
Y_valid = Y_valid_raw.copy()

for k, v in cats_map.iteritems():
    Y_train[Y_train_raw==k] = v
    Y_valid[Y_valid_raw==k] = v
```

```python
Y_t = Y_train.copy()
Y_v = Y_valid.copy()

Y_t[Y_train==0] = 0
Y_t[Y_train==1] = 1
Y_t[Y_train==4] = 2
Y_t[Y_train==6] = 3

Y_v[Y_valid==0] = 0
Y_v[Y_valid==1] = 1
Y_v[Y_valid==4] = 2
Y_v[Y_valid==6] = 3

categories_reduced = [0,1,2,3]

#Y_t_cat = to_categorical(Y_t)
#Y_v_cat = to_categorical(Y_v)
```
